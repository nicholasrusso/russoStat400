```python
import pandas as pd
from salmon import *
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv("tutorials/data/abalone/Dataset.data", 
                   names=["sex", "length", "diameter", "height", 
                           "whole_weight", "shucked_weight", 
                           "viscera_weight", "shell_weight", "rings"],
                  sep=" ")
# removing some outliers.
data = data[data["height"] < 0.4].reset_index(drop=True)
data.head(3)
```

```python
plt.scatter(data["length"], data["height"], 3)
plt.xlabel("length")
plt.ylabel("height")
plt.title("Abalone length vs height")
#plt.savefig('bad1.png', bbox_inches='tight')
```

```python
plt.scatter(data["length"], data["height"])
plt.xlabel("length")
plt.ylabel("height")
plt.plot([0.0, 0.4], [0.0, 0.3], color='k', linestyle='-', linewidth=2)
slope, intercept = np.polyfit([0.0, 0.4], [0.0, 0.3], 1)
plt.title("y = "+ str(slope) + "x + " + str(intercept))
#plt.title("Abalone length vs height")
plt.savefig('bad1.png', bbox_inches='tight')
```

```python
plt.scatter(data["length"], data["height"])
plt.xlabel("length")
plt.ylabel("height")
plt.plot([0.1, 0.7], [0.03, 0.1], color='k', linestyle='-', linewidth=2)
slope, intercept = np.polyfit([0.1, 0.7], [0.03, 0.1], 1)
plt.title("y = "+ str(round(slope, 2)) + "x + " + str(round(intercept, 2)))
#plt.title("Abalone length vs height")
plt.savefig('bad2.png', bbox_inches='tight')
```
